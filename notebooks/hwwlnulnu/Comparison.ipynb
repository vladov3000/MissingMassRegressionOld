```python
import pandas as pd
import seaborn
import numpy as np

pd.set_option('display.max_columns', None)
seaborn.set_theme()
```

```python
df = pd.read_pickle("../../hwwlnulnu/data/h_ww_lnulnu.pkl")
df = df.iloc[:, :1000]
```

```python
seaborn.displot(data={"Generated": df["H_Genm"], "Reconstructed": df["Hm"]}, binrange=[0, 2000], stat="density")
```

```python
data = {}

H_middles = [ 200. , 225. , 250. , 275. , 300. ]

for i in H_middles:
    data[str(i)] = df.loc[df.index[df["H_Mass"] == i], "Hm"]
```

```python
seaborn.displot(data=data, kind="kde", clip=[0, 750])
```
